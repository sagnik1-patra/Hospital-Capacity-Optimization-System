```python
import os
import json
import yaml
import pickle
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, classification_report, r2_score, mean_absolute_error

import tensorflow as tf
from tensorflow.keras.models import Sequential, model_from_json
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


# =========================
# PATHS
# =========================
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Hospital Capacity Optimization System"
CSV_PATH = os.path.join(
    BASE_DIR,
    "indian_medicine_head_quarters_hospitals_and_bed_strength_2019.csv"
)

os.makedirs(BASE_DIR, exist_ok=True)


# =========================
# LOAD DATA
# =========================
df = pd.read_csv(CSV_PATH)

print("Dataset Loaded Successfully")
print(df.head())
print(df.shape)
print(df.columns)

df.columns = df.columns.str.strip()

bed_col = "Sanctioned Bed Strength (in Numbers) as on 01.01.2018"
hospital_col = "Name of the Headquarters Hospital"

df[bed_col] = pd.to_numeric(df[bed_col], errors="coerce")
df[bed_col] = df[bed_col].fillna(df[bed_col].median())
df[hospital_col] = df[hospital_col].astype(str)


# =========================
# FEATURE ENGINEERING
# =========================
df["Doctors_Required"] = np.ceil(df[bed_col] / 20).astype(int)
df["Nurses_Required"] = np.ceil(df[bed_col] / 5).astype(int)
df["ICU_Beds_Required"] = np.ceil(df[bed_col] * 0.15).astype(int)
df["Ambulances_Required"] = np.ceil(df[bed_col] / 100).astype(int)
df["Emergency_Staff_Required"] = np.ceil(df[bed_col] / 25).astype(int)

df["Equipment_Score"] = (
    df["Doctors_Required"]
    + df["Nurses_Required"]
    + df["ICU_Beds_Required"]
    + df["Ambulances_Required"]
    + df["Emergency_Staff_Required"]
)


def capacity_category(beds):
    if beds <= 75:
        return "Small"
    elif beds <= 150:
        return "Medium"
    elif beds <= 300:
        return "Large"
    else:
        return "Very Large"


df["Capacity_Category"] = df[bed_col].apply(capacity_category)

print("\nClass Distribution:")
print(df["Capacity_Category"].value_counts())


# =========================
# ENCODING
# =========================
label_encoder = LabelEncoder()
df["Capacity_Label"] = label_encoder.fit_transform(df["Capacity_Category"])

features = [
    bed_col,
    "Doctors_Required",
    "Nurses_Required",
    "ICU_Beds_Required",
    "Ambulances_Required",
    "Emergency_Staff_Required",
    "Equipment_Score"
]

X = df[features]
y = df["Capacity_Label"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

num_classes = len(label_encoder.classes_)
y_cat = to_categorical(y, num_classes=num_classes)


# =========================
# SAFE TRAIN TEST SPLIT
# =========================
class_counts = df["Capacity_Label"].value_counts()

if class_counts.min() < 2:
    print("\nWarning: One class has only 1 sample.")
    print("Using normal train_test_split without stratify.")
    stratify_value = None
else:
    stratify_value = y

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_cat,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=stratify_value
)

y_train_labels = np.argmax(y_train, axis=1)
y_test_labels = np.argmax(y_test, axis=1)


# =========================
# ANN MODEL
# =========================
ann_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(num_classes, activation="softmax")
])

ann_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = ann_model.fit(
    X_train,
    y_train,
    epochs=100,
    batch_size=4,
    validation_split=0.2,
    verbose=1
)

loss, ann_accuracy = ann_model.evaluate(X_test, y_test, verbose=0)

print("\nANN Accuracy:", ann_accuracy)


# =========================
# RANDOM FOREST CLASSIFIER
# =========================
rf_classifier = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_classifier.fit(X_train, y_train_labels)

rf_pred = rf_classifier.predict(X_test)
rf_accuracy = accuracy_score(y_test_labels, rf_pred)

print("\nRandom Forest Accuracy:", rf_accuracy)

available_test_labels = sorted(np.unique(y_test_labels))

print(classification_report(
    y_test_labels,
    rf_pred,
    labels=available_test_labels,
    target_names=label_encoder.inverse_transform(available_test_labels),
    zero_division=0
))


# =========================
# REGRESSION MODEL
# =========================
regression_targets = [
    "Doctors_Required",
    "Nurses_Required",
    "ICU_Beds_Required",
    "Ambulances_Required",
    "Emergency_Staff_Required"
]

X_reg = df[[bed_col]]
y_reg = df[regression_targets]

reg_scaler = StandardScaler()
X_reg_scaled = reg_scaler.fit_transform(X_reg)

Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_reg_scaled,
    y_reg,
    test_size=0.2,
    random_state=42
)

rf_regressor = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_regressor.fit(Xr_train, yr_train)

yr_pred = rf_regressor.predict(Xr_test)

print("\nRegression R2 Score:", r2_score(yr_test, yr_pred))
print("Regression MAE:", mean_absolute_error(yr_test, yr_pred))


# =========================
# SAVE H5 MODEL
# =========================
h5_path = os.path.join(BASE_DIR, "hospital_capacity_ann_model.h5")
ann_model.save(h5_path)
print("H5 model saved:", h5_path)


# =========================
# SAVE JSON MODEL
# =========================
json_model_path = os.path.join(BASE_DIR, "hospital_capacity_ann_model.json")

with open(json_model_path, "w") as json_file:
    json_file.write(ann_model.to_json())

print("JSON model saved:", json_model_path)


# =========================
# SAVE YAML CONFIG
# =========================
yaml_model_path = os.path.join(BASE_DIR, "hospital_capacity_ann_model.yaml")

model_config = {
    "project_name": "Smart Healthcare Resource Allocation and Hospital Capacity Optimization System",
    "model_type": "ANN + Random Forest",
    "input_features": features,
    "target": "Capacity_Category",
    "classes": label_encoder.classes_.tolist(),
    "ann_accuracy": float(ann_accuracy),
    "random_forest_accuracy": float(rf_accuracy),
    "epochs": 100,
    "batch_size": 4,
    "note": "Stratify disabled automatically if any class has fewer than 2 samples."
}

with open(yaml_model_path, "w") as yaml_file:
    yaml.dump(model_config, yaml_file, sort_keys=False)

print("YAML config saved:", yaml_model_path)


# =========================
# SAVE PKL PACKAGE
# =========================
pkl_path = os.path.join(BASE_DIR, "hospital_capacity_models.pkl")

model_package = {
    "ann_model_path": h5_path,
    "rf_classifier": rf_classifier,
    "rf_regressor": rf_regressor,
    "label_encoder": label_encoder,
    "classification_scaler": scaler,
    "regression_scaler": reg_scaler,
    "features": features,
    "regression_targets": regression_targets
}

with open(pkl_path, "wb") as pkl_file:
    pickle.dump(model_package, pkl_file)

print("PKL model package saved:", pkl_path)


# =========================
# SAVE PROCESSED DATA
# =========================
processed_csv_path = os.path.join(BASE_DIR, "hospital_capacity_processed_data.csv")
df.to_csv(processed_csv_path, index=False)

print("Processed CSV saved:", processed_csv_path)


# =========================
# SAVE PREDICTIONS
# =========================
ann_pred_probs = ann_model.predict(X_scaled)
ann_pred_labels = np.argmax(ann_pred_probs, axis=1)
ann_pred_categories = label_encoder.inverse_transform(ann_pred_labels)

df["Predicted_Capacity_Category"] = ann_pred_categories

resource_predictions = rf_regressor.predict(reg_scaler.transform(df[[bed_col]]))

for i, col in enumerate(regression_targets):
    df["Predicted_" + col] = np.round(resource_predictions[:, i]).astype(int)

prediction_csv_path = os.path.join(BASE_DIR, "hospital_capacity_predictions.csv")
df.to_csv(prediction_csv_path, index=False)

print("Prediction CSV saved:", prediction_csv_path)


# =========================
# MODEL RELOAD TEST
# =========================
with open(json_model_path, "r") as json_file:
    loaded_model_json = json_file.read()

loaded_model = model_from_json(loaded_model_json)
loaded_model.load_weights(h5_path)

print("\nModel reload test successful.")


print("\nAll files generated successfully in:")
print(BASE_DIR)
```
