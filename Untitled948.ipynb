```python
import os
import json
import yaml
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    r2_score,
    mean_absolute_error,
    mean_squared_error
)

import tensorflow as tf
from tensorflow.keras.models import Sequential, model_from_json
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


# =========================
# PATHS
# =========================
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Hospital Capacity Optimization System"
CSV_PATH = os.path.join(
    BASE_DIR,
    "indian_medicine_head_quarters_hospitals_and_bed_strength_2019.csv"
)

os.makedirs(BASE_DIR, exist_ok=True)


# =========================
# LOAD DATA
# =========================
df = pd.read_csv(CSV_PATH)
df.columns = df.columns.str.strip()

bed_col = "Sanctioned Bed Strength (in Numbers) as on 01.01.2018"
hospital_col = "Name of the Headquarters Hospital"

df[bed_col] = pd.to_numeric(df[bed_col], errors="coerce")
df[bed_col] = df[bed_col].fillna(df[bed_col].median())
df[hospital_col] = df[hospital_col].astype(str)

print("Dataset Loaded Successfully")
print(df.head())
print(df.shape)


# =========================
# FEATURE ENGINEERING
# =========================
df["Doctors_Required"] = np.ceil(df[bed_col] / 20).astype(int)
df["Nurses_Required"] = np.ceil(df[bed_col] / 5).astype(int)
df["ICU_Beds_Required"] = np.ceil(df[bed_col] * 0.15).astype(int)
df["Ambulances_Required"] = np.ceil(df[bed_col] / 100).astype(int)
df["Emergency_Staff_Required"] = np.ceil(df[bed_col] / 25).astype(int)

df["Equipment_Score"] = (
    df["Doctors_Required"]
    + df["Nurses_Required"]
    + df["ICU_Beds_Required"]
    + df["Ambulances_Required"]
    + df["Emergency_Staff_Required"]
)

def capacity_category(beds):
    if beds <= 75:
        return "Small"
    elif beds <= 150:
        return "Medium"
    elif beds <= 300:
        return "Large"
    else:
        return "Very Large"

df["Capacity_Category"] = df[bed_col].apply(capacity_category)

print("\nCapacity Class Distribution:")
print(df["Capacity_Category"].value_counts())


# =========================
# ENCODING
# =========================
label_encoder = LabelEncoder()
df["Capacity_Label"] = label_encoder.fit_transform(df["Capacity_Category"])

features = [
    bed_col,
    "Doctors_Required",
    "Nurses_Required",
    "ICU_Beds_Required",
    "Ambulances_Required",
    "Emergency_Staff_Required",
    "Equipment_Score"
]

X = df[features]
y = df["Capacity_Label"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

num_classes = len(label_encoder.classes_)
y_cat = to_categorical(y, num_classes=num_classes)


# =========================
# SAFE TRAIN TEST SPLIT
# =========================
if df["Capacity_Label"].value_counts().min() < 2:
    stratify_value = None
else:
    stratify_value = y

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_cat,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=stratify_value
)

y_train_labels = np.argmax(y_train, axis=1)
y_test_labels = np.argmax(y_test, axis=1)


# =========================
# ANN MODEL
# =========================
ann_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(num_classes, activation="softmax")
])

ann_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = ann_model.fit(
    X_train,
    y_train,
    epochs=100,
    batch_size=4,
    validation_split=0.2,
    verbose=1
)

ann_loss, ann_accuracy = ann_model.evaluate(X_test, y_test, verbose=0)
ann_pred = np.argmax(ann_model.predict(X_test), axis=1)


# =========================
# RANDOM FOREST CLASSIFIER
# =========================
rf_classifier = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_classifier.fit(X_train, y_train_labels)
rf_pred = rf_classifier.predict(X_test)
rf_accuracy = accuracy_score(y_test_labels, rf_pred)


# =========================
# DECISION TREE CLASSIFIER
# =========================
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train_labels)
dt_pred = dt_classifier.predict(X_test)
dt_accuracy = accuracy_score(y_test_labels, dt_pred)


# =========================
# REGRESSION MODEL
# =========================
regression_targets = [
    "Doctors_Required",
    "Nurses_Required",
    "ICU_Beds_Required",
    "Ambulances_Required",
    "Emergency_Staff_Required"
]

X_reg = df[[bed_col]]
y_reg = df[regression_targets]

reg_scaler = StandardScaler()
X_reg_scaled = reg_scaler.fit_transform(X_reg)

Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_reg_scaled,
    y_reg,
    test_size=0.2,
    random_state=42
)

rf_regressor = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_regressor.fit(Xr_train, yr_train)
yr_pred = rf_regressor.predict(Xr_test)

reg_r2 = r2_score(yr_test, yr_pred)
reg_mae = mean_absolute_error(yr_test, yr_pred)
reg_rmse = np.sqrt(mean_squared_error(yr_test, yr_pred))


# =========================
# SAVE MODELS
# =========================
h5_path = os.path.join(BASE_DIR, "hospital_capacity_ann_model.h5")
json_path = os.path.join(BASE_DIR, "hospital_capacity_ann_model.json")
yaml_path = os.path.join(BASE_DIR, "hospital_capacity_ann_model.yaml")
pkl_path = os.path.join(BASE_DIR, "hospital_capacity_models.pkl")

ann_model.save(h5_path)

with open(json_path, "w") as f:
    f.write(ann_model.to_json())

config = {
    "project_name": "Smart Healthcare Resource Allocation and Hospital Capacity Optimization System",
    "model_type": "ANN, Random Forest, Decision Tree",
    "features": features,
    "target": "Capacity_Category",
    "classes": label_encoder.classes_.tolist(),
    "ann_accuracy": float(ann_accuracy),
    "random_forest_accuracy": float(rf_accuracy),
    "decision_tree_accuracy": float(dt_accuracy),
    "regression_r2": float(reg_r2),
    "regression_mae": float(reg_mae),
    "regression_rmse": float(reg_rmse)
}

with open(yaml_path, "w") as f:
    yaml.dump(config, f, sort_keys=False)

model_package = {
    "ann_model_path": h5_path,
    "rf_classifier": rf_classifier,
    "dt_classifier": dt_classifier,
    "rf_regressor": rf_regressor,
    "label_encoder": label_encoder,
    "classification_scaler": scaler,
    "regression_scaler": reg_scaler,
    "features": features,
    "regression_targets": regression_targets
}

with open(pkl_path, "wb") as f:
    pickle.dump(model_package, f)


# =========================
# RESULT CSV
# =========================
results_df = pd.DataFrame({
    "Model": [
        "ANN Classifier",
        "Random Forest Classifier",
        "Decision Tree Classifier",
        "Random Forest Regressor"
    ],
    "Accuracy": [
        ann_accuracy,
        rf_accuracy,
        dt_accuracy,
        np.nan
    ],
    "R2_Score": [
        np.nan,
        np.nan,
        np.nan,
        reg_r2
    ],
    "MAE": [
        np.nan,
        np.nan,
        np.nan,
        reg_mae
    ],
    "RMSE": [
        np.nan,
        np.nan,
        np.nan,
        reg_rmse
    ]
})

result_csv_path = os.path.join(BASE_DIR, "hospital_capacity_result.csv")
results_df.to_csv(result_csv_path, index=False)


# =========================
# PREDICTION CSV
# =========================
full_pred_probs = ann_model.predict(X_scaled)
full_pred_labels = np.argmax(full_pred_probs, axis=1)
full_pred_categories = label_encoder.inverse_transform(full_pred_labels)

df["Predicted_Capacity_Category"] = full_pred_categories

resource_predictions = rf_regressor.predict(reg_scaler.transform(df[[bed_col]]))

for i, col in enumerate(regression_targets):
    df["Predicted_" + col] = np.round(resource_predictions[:, i]).astype(int)

prediction_csv_path = os.path.join(BASE_DIR, "hospital_capacity_prediction.csv")
df.to_csv(prediction_csv_path, index=False)


# =========================
# ACCURACY GRAPH
# =========================
plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("ANN Training and Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
accuracy_graph_path = os.path.join(BASE_DIR, "hospital_capacity_accuracy_graph.png")
plt.savefig(accuracy_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# =========================
# HEATMAP / CONFUSION MATRIX
# =========================
cm = confusion_matrix(y_test_labels, rf_pred)

plt.figure(figsize=(7, 5))
plt.imshow(cm, cmap="Blues")
plt.title("Random Forest Confusion Matrix Heatmap")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.colorbar()

classes = label_encoder.classes_
plt.xticks(range(len(classes)), classes, rotation=45)
plt.yticks(range(len(classes)), classes)

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.tight_layout()
heatmap_path = os.path.join(BASE_DIR, "hospital_capacity_heatmap.png")
plt.savefig(heatmap_path, dpi=300, bbox_inches="tight")
plt.show()


# =========================
# COMPARISON GRAPH
# =========================
comparison_df = results_df.dropna(subset=["Accuracy"])

plt.figure(figsize=(8, 5))
plt.bar(comparison_df["Model"], comparison_df["Accuracy"])
plt.title("Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=25)
plt.ylim(0, 1)
plt.grid(axis="y")
comparison_graph_path = os.path.join(BASE_DIR, "hospital_capacity_comparison_graph.png")
plt.savefig(comparison_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# =========================
# RESULT GRAPH
# =========================
plt.figure(figsize=(8, 5))
plt.bar(results_df["Model"], results_df["Accuracy"].fillna(0))
plt.title("Model Result Graph")
plt.xlabel("Model")
plt.ylabel("Score")
plt.xticks(rotation=25)
plt.grid(axis="y")
result_graph_path = os.path.join(BASE_DIR, "hospital_capacity_result_graph.png")
plt.savefig(result_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# =========================
# PREDICTION GRAPH
# =========================
plt.figure(figsize=(12, 6))
plt.bar(df[hospital_col], df[bed_col])
plt.title("Hospital Bed Strength Prediction Graph")
plt.xlabel("Hospital")
plt.ylabel("Sanctioned Bed Strength")
plt.xticks(rotation=90)
plt.grid(axis="y")
prediction_graph_path = os.path.join(BASE_DIR, "hospital_capacity_prediction_graph.png")
plt.savefig(prediction_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# =========================
# RESOURCE ALLOCATION GRAPH
# =========================
resource_sum = df[
    [
        "Doctors_Required",
        "Nurses_Required",
        "ICU_Beds_Required",
        "Ambulances_Required",
        "Emergency_Staff_Required"
    ]
].sum()

plt.figure(figsize=(8, 5))
plt.bar(resource_sum.index, resource_sum.values)
plt.title("Total Healthcare Resource Allocation")
plt.xlabel("Resource Type")
plt.ylabel("Total Required")
plt.xticks(rotation=30)
plt.grid(axis="y")
resource_graph_path = os.path.join(BASE_DIR, "hospital_capacity_resource_allocation_graph.png")
plt.savefig(resource_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# =========================
# SAVE CLASSIFICATION REPORT
# =========================
report = classification_report(
    y_test_labels,
    rf_pred,
    labels=np.unique(y_test_labels),
    target_names=label_encoder.inverse_transform(np.unique(y_test_labels)),
    zero_division=0,
    output_dict=True
)

report_df = pd.DataFrame(report).transpose()
report_csv_path = os.path.join(BASE_DIR, "hospital_capacity_classification_report.csv")
report_df.to_csv(report_csv_path)


# =========================
# SAVE PROCESSED DATA
# =========================
processed_csv_path = os.path.join(BASE_DIR, "hospital_capacity_processed_data.csv")
df.to_csv(processed_csv_path, index=False)


# =========================
# FINAL OUTPUT
# =========================
print("\nAll results saved successfully!")

print("\nSaved Files:")
print(h5_path)
print(json_path)
print(yaml_path)
print(pkl_path)
print(result_csv_path)
print(prediction_csv_path)
print(report_csv_path)
print(processed_csv_path)
print(accuracy_graph_path)
print(heatmap_path)
print(comparison_graph_path)
print(result_graph_path)
print(prediction_graph_path)
print(resource_graph_path)
```
